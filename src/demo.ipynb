```python


class ExamQuestion(BaseModel):
    '''è€ƒè¯•é¢˜ç›®ä¿¡æ¯'''
    question: str = Field(description="è€ƒè¯•é¢˜ç›®")
    answer: str = Field(description="è€ƒè¯•é¢˜ç›®ç­”æ¡ˆ")
    question_type: str = Field(description="è€ƒè¯•é¢˜ç›®ç±»å‹")
    knowledge_point: str = Field(description="è€ƒè¯•é¢˜ç›®æ‰€å±çš„çŸ¥è¯†ç‚¹")
    knowledge_point_description: str = Field(description="è€ƒè¯•é¢˜ç›®æ‰€å±çš„çŸ¥è¯†ç‚¹çš„å…·ä½“æè¿°")
    extra_requirement: Optional[str] = Field(description="è€ƒè¯•é¢˜ç›®é¢å¤–è¦æ±‚")

exam_question = ExamQuestion(
    question='''
    æœç´¢ç®—æ³•ç›¸å…³\nï¼ˆ1ï¼‰åˆ†åˆ«è¯´æ˜ DFS å’Œ BFS å¦‚ä½•ç”¨é˜Ÿåˆ—æˆ–æ ˆå®ç°ï¼Œå¹¶å¯¹æ¯”ä¸¤è€…éå†åŒä¸€å›¾æ—¶çš„é¡ºåºå·®å¼‚ã€‚\nï¼ˆ2ï¼‰åœ¨æ±‚è§£æ— æƒå›¾æœ€çŸ­è·¯å¾„é—®é¢˜æ—¶ï¼Œä¸ºä»€ä¹ˆ BFS é€šå¸¸æ¯” DFS æ›´é«˜æ•ˆï¼Ÿç»“åˆéå†ç‰¹æ€§è§£é‡ŠåŸå› ã€‚
    ''',
    answer="ï¼ˆ1ï¼‰DFS ç”¨æ ˆï¼ˆé€’å½’æˆ–æ˜¾å¼æ ˆï¼‰ï¼Œä¸€è·¯æ·±å…¥å†å›æº¯ï¼›BFS ç”¨é˜Ÿåˆ—ï¼Œä¸€å±‚å±‚æ‰©å±•ï¼›é¡ºåºå·®å¼‚ï¼šDFS çºµæ·±ï¼ŒBFS æ¨ªæ‰©ã€‚\nï¼ˆ2ï¼‰BFS æŒ‰å±‚æ‰©å±•ï¼Œé¦–æ¬¡åˆ°è¾¾ç›®æ ‡å³æœ€çŸ­è·¯å¾„ï¼›DFS å¯èƒ½æ·±å…¥å¾ˆé•¿éæœ€çŸ­è·¯å¾„æ‰å›æº¯ï¼Œè®¿é—®èŠ‚ç‚¹æ›´å¤šã€‚",
    question_type="ç®€ç­”é¢˜",
    knowledge_point="",
    knowledge_point_description="",
    extra_requirement="å°†ç®€ç­”é¢˜ä¿®æ”¹ä¸ºå•é€‰é¢˜",
)

text = '''
question:{question}
answer:{answer}
question_type:{question_type}
knowledge_point:{knowledge_point}
knowledge_point_description:{knowledge_point_description}
extra_requirement:{extra_requirement}
'''
inputs = text.format(**exam_question.model_dump())

def test_deployed_agent():
    # å‡†å¤‡æµ‹è¯•è´Ÿè½½
    payload = {
        "input": [
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": inputs},
                ],
            },
        ],
        "session_id": "test_session_001",
        "user_id": "test_user_001",
    }

    print("ğŸ§ª æµ‹è¯•éƒ¨ç½²çš„æ™ºèƒ½ä½“...")

    # æµ‹è¯•æµå¼å“åº”
    try:
        response = requests.post(
            "http://localhost:8001/process",
            json=payload,
            stream=True,
            timeout=30,
        )
        res = []
        # é€è¡Œè§£æ SSE æ•°æ®å¹¶å³æ—¶æ‰“å°
        for line in response.iter_lines(decode_unicode=True):
            if not line:
                continue

            # ä»…å¤„ç† data: å¼€å¤´çš„è¡Œ
            if line.startswith("data:"):
                s = line.split(":", 1)[1].strip()
                if s == "[DONE]":
                    break

                # å°è¯•è§£æ JSONï¼›å¤±è´¥æ—¶ç›´æ¥åŸæ ·æ‰“å°
                try:
                    data = json.loads(s)
                except json.JSONDecodeError:
                    print(s, end="", flush=True)
                    continue

                obj = data.get("object")
                if obj == "content":
                    text = data.get("text", "")
                    if text:
                        print(text, end="", flush=True)
                        res.append(text)
                elif obj == "status":
                    st = data.get("status") or data.get("text") or ""
                    if st:
                        print(f"\n[çŠ¶æ€] {st}\n", end="", flush=True)
                elif obj == "error":
                    print(f"\n[é”™è¯¯] {data.get('text') or data}\n", flush=True)
                    break
        
    except requests.exceptions.RequestException as e:
        print(f"âŒ æµå¼æµ‹è¯•å¤±è´¥: {e}")
    except requests.exceptions.RequestException as e:
        print(f"â„¹ï¸ JSONç«¯ç‚¹ä¸å¯ç”¨æˆ–å¤±è´¥: {e}")


# Run the test
test_deployed_agent()
```
